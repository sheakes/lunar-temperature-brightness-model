# Spherical Plotting

```python
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Suppose your data is shape (360, 180) for (longitude, latitude)
# and your longitude range is [-180, 180], latitude range is [-90, 90].
# data[lon_idx, lat_idx]
data = np.random.rand(360, 180)  # placeholder

# Generate coordinate arrays
lon = np.linspace(-180, 180, 360)
lat = np.linspace(-90, 90, 180)
Lon, Lat = np.meshgrid(lon, lat, indexing='ij')

# Set up a map
fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection=ccrs.PlateCarree())

# Optionally set the extent (lon_min, lon_max, lat_min, lat_max)
ax.set_extent([-180, 180, -90, 90], ccrs.PlateCarree())

# Plot the data
# One method: pcolormesh handles 2D lat–lon data
# Note: pass Lon, Lat in, and the data array must match that shape
mesh = ax.pcolormesh(Lon, Lat, data, 
                     cmap='viridis', 
                     shading='auto', 
                     transform=ccrs.PlateCarree())

# Add features
ax.coastlines()
ax.gridlines(draw_labels=True)

# Colorbar
cbar = plt.colorbar(mesh, orientation='vertical', pad=0.02)
cbar.set_label('Surface Temperature (K)')

plt.title("Lunar Surface Temperature")
plt.show()
```
